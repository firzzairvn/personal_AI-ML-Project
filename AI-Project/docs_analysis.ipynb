````python
# Extract Text
import pdfplumber

pdf_source  = '/content/Sejarah Batik.pdf'
text_source = 'Sejarah_Batik.txt'
with pdfplumber.open(pdf_source) as pdf:
    text = ""
    for page in pdf.pages:
        text += page.extract_text()

with open(text_source, "w") as text_file:
    text_file.write(text)

print(f"Text extracted and saved to {text_source}")
````

````python
# Read the extracted text
with open("/content/Sejarah_Batik.txt", "r") as file:
    document_text = file.read()

print(document_text[:1000])  # max length is 1000
````

````python
from transformers import pipeline

# Inisialisasi pipeline untuk summarization dengan model T5-small
summarizer = pipeline("summarization", model="t5-small")

# Fungsi untuk membagi teks besar menjadi bagian-bagian yang lebih kecil
def split_text(text, max_length=1000):
    return [text[i:i+max_length] for i in range(0, len(text), max_length)]

if len(document_text) > 1000:
    chunks = split_text(document_text, max_length=1000)
    summaries = []
    for chunk in chunks:
        # Pastikan chunk tidak kosong dan pangkas spasi ekstra
        chunk = chunk.strip()
        if chunk:
            summary = summarizer(chunk, max_length=150, min_length=30, do_sample=False)
            summaries.append(summary[0]['summary_text'])
    # Menggabungkan ringkasan tiap bagian
    final_summary = "\n".join(summaries)
else:
    summary = summarizer(document_text, max_length=150, min_length=30, do_sample=False)
    final_summary = summary[0]['summary_text']

print("Summary:\n", final_summary)
````

## Penjelasan Kode Summarization  

Kode ini melakukan **proses peringkasan (summarization)** pada teks yang diambil dari file `.txt` menggunakan model **`t5-small`** dari library **Hugging Face Transformers**.  

### 1. Membaca File Teks  
- Kode membaca seluruh isi file teks dengan **encoding `utf-8`** untuk memastikan karakter **non-ASCII** terbaca dengan benar.  

### 2. Inisialisasi Model Summarization  
- **Pipeline summarization** diinisialisasi dengan model `t5-small`.  

### 3. Penanganan Teks Panjang  
- Karena model memiliki batasan **panjang input**, kode akan **memeriksa panjang teks**.  
- Jika teks lebih dari **1000 karakter**, maka teks akan dipecah menjadi beberapa bagian menggunakan fungsi **`split_text`**.  

### 4. Proses Peringkasan  
- Setiap potongan teks dirangkum secara terpisah menggunakan **pipeline summarization**.  
- Parameter yang digunakan dalam summarization:  
  - **`max_length`** → Menentukan panjang maksimum ringkasan.  
  - **`min_length`** → Menentukan panjang minimum ringkasan.  

Kode ini memastikan bahwa teks panjang tetap dapat dirangkum dengan baik tanpa melebihi batas input model.  


````python
import nltk
nltk.download('punkt')
nltk.download('punkt_tab')
from nltk.tokenize import sent_tokenize
````

````python
def split_text_into_passages(text, max_words=200):
    sentences = sent_tokenize(text)
    passages, current_passage = [], []

    for sentence in sentences:
        if sum(len(s.split()) for s in current_passage) + len(sentence.split()) < max_words:
            current_passage.append(sentence)
        else:
            passages.append(" ".join(current_passage).strip())
            current_passage = [sentence]

    if current_passage:
        passages.append(" ".join(current_passage).strip())

    return passages

passages = split_text_into_passages(final_summary, max_words=50)
````

## Penjelasan Kode: Pembagian Teks Menjadi Passages

Kode di bawah ini bertujuan untuk membagi sebuah teks panjang menjadi beberapa bagian (passages) berdasarkan batas jumlah kata. Fungsi ini sangat berguna untuk memproses teks dalam aplikasi NLP, seperti summarization atau question generation, di mana pemrosesan teks dalam potongan kecil seringkali lebih efisien dan sesuai dengan batasan model.

### Fungsi `split_text_into_passages`

```python
def split_text_into_passages(text, max_words=200):

````python
qg_pipeline = pipeline("text2text-generation", model="valhalla/t5-base-qg-hl")

def generate_questions_pipeline(passage: str, min_questions: int = 3) -> list:
    input_text = f"generate questions: {passage}"
    results = qg_pipeline(input_text)
    questions = [q.strip() for q in results[0]['generated_text'].split('<sep>') if q.strip()]

    if len(questions) < min_questions:
        passage_sentences = passage.split('. ')
        for i in range(len(passage_sentences)):
            if len(questions) >= min_questions:
                break
            additional_input = ' '.join(passage_sentences[i:i+2])
            additional_results = qg_pipeline(f"generate questions: {additional_input}")
            additional_questions = [q.strip() for q in additional_results[0]['generated_text'].split('<sep>') if q.strip()]
            questions.extend(additional_questions)
    return questions[:min_questions]

for idx, passage in enumerate(passages, start=1):
    questions = generate_questions_pipeline(passage)
    print(f"Passage {idx}:\n{passage}\n")
    print("Generated Questions:")
    for q in questions:
        print(f"- {q}")
    print("\n" + "-" * 50 + "\n")
````

## Penjelasan Kode

Kode ini digunakan untuk menghasilkan pertanyaan dari sebuah teks menggunakan model **T5** yang telah dilatih untuk tugas **question generation**.

### Fungsi dan Peran Kode:
1. **Inisialisasi Model (`qg_pipeline`)**  
   - Menggunakan model `"valhalla/t5-base-qg-hl"` untuk menghasilkan pertanyaan berdasarkan teks masukan.

2. **Fungsi `generate_questions_pipeline(passage, min_questions=3)`**  
   - Menerima teks (`passage`) dan jumlah minimum pertanyaan yang diinginkan (`min_questions`).
   - Menghasilkan pertanyaan berdasarkan teks menggunakan `qg_pipeline`.
   - Jika jumlah pertanyaan kurang dari `min_questions`, maka teks dipecah menjadi kalimat, dan pertanyaan tambahan dibuat dari gabungan dua kalimat.

3. **Looping Melalui `passages`**  
   - Untuk setiap teks dalam `passages`, fungsi `generate_questions_pipeline()` dipanggil untuk menghasilkan pertanyaan.
   - Hasilnya ditampilkan dalam format yang rapi.

````python
from transformers import pipeline
qa_pipeline = pipeline("question-answering", model="deepset/roberta-base-squad2")

def answer_unique_questions(passages: list, qa_pipeline) -> None:
    answered_questions = set()

    for idx, passage in enumerate(passages, start=1):
        questions = generate_questions_pipeline(passage)
        print(f"Passage {idx}:\n{passage}\n")

        for question in questions:
            if question not in answered_questions:
                qa_input = {
                    'question': question,
                    'context': passage
                }
                answer = qa_pipeline(qa_input)
                print(f"Q: {question}")
                print(f"A: {answer['answer']}\n")
                answered_questions.add(question)

        print("=" * 50, "\n")
answer_unique_questions(passages, qa_pipeline)
````

## Penjelasan Kode

Kode ini menggabungkan fungsi **question generation** dan **question answering** untuk menjawab pertanyaan unik yang dihasilkan dari masing-masing teks.

### Fungsi dan Peran Kode:
1. **Inisialisasi Model Question Answering (`qa_pipeline`)**  
   - Menggunakan model `"deepset/roberta-base-squad2"` dari Hugging Face untuk menjawab pertanyaan berdasarkan konteks yang diberikan.

2. **Fungsi `answer_unique_questions(passages, qa_pipeline)`**  
   - Menerima daftar teks (`passages`) dan pipeline QA.
   - Menggunakan fungsi `generate_questions_pipeline` (didefinisikan sebelumnya) untuk menghasilkan pertanyaan dari setiap teks.
   - Menyimpan pertanyaan yang telah dijawab agar tidak terjadi duplikasi.
   - Untuk setiap pertanyaan unik, kode membuat input berupa dictionary yang berisi `question` dan `context` (teks yang relevan) kemudian memanggil `qa_pipeline` untuk mendapatkan jawaban.
   - Hasilnya ditampilkan dengan format yang rapi, mencakup teks, pertanyaan, dan jawabannya.

3. **Eksekusi Fungsi**  
   - Memanggil fungsi `answer_unique_questions` dengan daftar teks `passages` dan pipeline QA untuk menampilkan hasil pertanyaan dan jawaban.
