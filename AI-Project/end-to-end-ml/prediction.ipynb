# About Dataset

## Car Price Prediction Dataset Description

This dataset contains **10,000 entries** created for the purpose of predicting car prices.  
Each row represents **information about a car and its price**. 

Dataset Source : https://www.kaggle.com/datasets/asinow/car-price-dataset/data

## Column Descriptions

### **1. Brand**  
Specifies the brand of the car (e.g., Toyota, BMW, Ford).  
**Example values:** `"Toyota"`, `"BMW"`, `"Mercedes"`

### **2. Model**  
Specifies the model of the car (e.g., Corolla, Focus, X5).  
**Example values:** `"Corolla"`, `"Focus"`, `"X5"`

### **3. Year**  
The production year of the car. Newer years typically indicate higher prices.  
**Example values:** `2005`, `2018`, `2023`

### **4. Engine_Size**  
Specifies the engine size in liters (L). Larger engines generally correlate with higher prices.  
**Example values:** `1.6`, `2.0`, `3.5`

### **5. Fuel_Type**  
Indicates the type of fuel used by the car:  
- **Petrol**: Cars running on gasoline.  
- **Diesel**: Cars running on diesel fuel.  
- **Hybrid**: Cars that use both fuel and electricity.  
- **Electric**: Fully electric cars.

### **6. Transmission**  
The type of transmission in the car:  
- **Manual**: Manual transmission.  
- **Automatic**: Automatic transmission.  
- **Semi-Automatic**: Semi-automatic transmission.  

### **7. Mileage**  
The total distance the car has traveled, measured in kilometers.  
Lower mileage generally indicates a higher price.  
**Example values:** `15,000`, `75,000`, `230,000`

### **8. Doors**  
The number of doors in the car. Commonly **2, 3, 4, or 5** doors.  
**Example values:** `2`, `3`, `4`, `5`

### **9. Owner_Count**  
The number of previous owners of the car.  
Fewer owners generally indicate a higher price.  
**Example values:** `1`, `2`, `3`, `4`

### **10. Price**  
The estimated selling price of the car.  
It is calculated based on several factors such as **production year, engine size, mileage, fuel type, and transmission**.  
**Example values:** `5,000`, `15,000`, `30,000`

### **Outlier Check & Remove** 

```python
plt.figure(figsize=(16,5))
plt.subplot(1,2,1)
sns.distplot(df['Price'])
plt.subplot(1,2,2)
sns.distplot(df['Mileage'])
plt.show()
```

```python
Q1 = df[['Mileage', 'Price']].quantile(0.25) #percentile 25
Q3 = df[['Mileage', 'Price']].quantile(0.75) #percentile 75
IQR = Q3 - Q1
```

```python
upper_limit = Q3 + 1.5 * IQR
lower_limit = Q1 - 1.5 * IQR
```

```python
df[df[['Mileage', 'Price']] > upper_limit]
df[df[['Mileage', 'Price']] < lower_limit]
```

```python
outliers = ((df[['Mileage', 'Price']] < lower_limit) | (df[['Mileage', 'Price']] > upper_limit)).sum()
outliers
```

```python
df_cleaned = df[(df['Price'] >= lower_limit['Price']) & (df['Price'] <= upper_limit['Price'])]
```

```python
df_cleaned.shape
```

```python
plt.figure(figsize=(16,8))
plt.subplot(2,2,1)
sns.distplot(df['Price'])
plt.subplot(2,2,2)
sns.boxplot(df['Mileage'])
plt.subplot(2,2,3)
sns.distplot(df_cleaned['Price'])
plt.subplot(2,2,4)
sns.boxplot(df_cleaned['Mileage'])
plt.show()
```

### **Build Model Linear Regression** 

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
```

```python
df_cleaned = df_cleaned.drop(columns=['Brand', 'Model', 'Fuel_Type', 'Transmission', 'Owner_Count'], axis=1)
```

```python
X = df_cleaned.drop('Price', axis=1)
y = df_cleaned['Price']
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)
model = LinearRegression()
```

```python
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
predictions_df = pd.DataFrame({'Actual Price': y_test.values, 'Predicted Price': y_pred})
print(predictions_df.head())  
```

```python
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

print(f"Mean Squared Error (MSE): {mse:.2f}")
print(f"R-squared (RÂ²): {r2:.2f}")
```

```python
new_data = [[2022, 2.0, 50000, 4]]  # Contoh input: Tahun, Engine Size, Mileage, Doors
new_data_scaled = scaler.transform(new_data)  # Standarisasi data

predicted_price = model.predict(new_data_scaled)
print(f"Predicted Price: ${predicted_price[0]:,.2f}")
```
